開啟分析資料檔，並進行資料預處理

```
import pandas as pd
df=pd.read_excel('loan3(na).xlsx')
```

```
df.head(5)
#loan=2表示違約
```

```
df.isnull().any() #尋找空值變數
```

```
df1=df.dropna()
X=df1[['Industry_type','Company_type','Beginning_time','Bank_relationship','Capital_increase','Age','Education','Credit']]
y=df1['loan'] 
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

判別分析語法

```
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
LDA.fit(X_train,y_train)
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
LDA_tpred= LDA.predict(X_train)
print(confusion_matrix(y_train,LDA_tpred))
print("======================================================\n")
print(classification_report(y_train,LDA_tpred))
```

```
#列印測試資料的混淆矩陣(confusion matrix)以及模式衡量指標
LDA_pred= LDA.predict(X_test)
print(confusion_matrix(y_test,LDA_pred))
print("======================================================\n")
print(classification_report(y_test,LDA_pred))
```

```
#輸入自變數數值，計算所預測之分類結果
LDA.predict([[1,1,1,3,1,4,3,1]])
```

```
#輸入自變數數值，計算所預測不同類別發生機率
LDA.predict_proba([[1,1,1,3,1,4,3,1]])
```

```
#同時輸入兩組自變數數值，計算所預測之分類結果
LDA.predict([[1,1,1,3,1,4,3,2],[1,1,1,3,1,4,3,1]])
```

```
#同時輸入兩組自變數數值，計算所預測不同類別發生機率
LDA.predict_proba([[1,1,1,3,1,4,3,2],[1,1,1,3,1,4,3,1]])
```

SVM 分析

```
from sklearn import svm 
clf = svm.SVC(kernel = 'linear') 
clf.fit(X_train,y_train) 
```

```
#列印訓練完成之模式的混淆矩陣(confusion matrix)以及模式衡量指標
from sklearn.metrics import classification_report,confusion_matrix
clf_tpred= clf.predict(X_train)
print(confusion_matrix(y_train,clf_tpred))
print("======================================================\n")
print(classification_report(y_train,clf_tpred))
```

```
clf_pred= clf.predict(X_test)
print(confusion_matrix(y_test,clf_pred))
print("======================================================\n")
print(classification_report(y_test,clf_pred))
```
