```python
import pandas as pd
df=pd.read_excel('loan4.xlsx')
```

```python
df.head()
```

```python
#將含有缺失值的列（特徵向量）去掉; 決定特徵變數(X)與目標變數(Y)
df1=df.dropna()
X=df1[['Industry_type','Company_type','Beginning_time','Bank_relationship','Capital_increase','Age','Education','Credit']]
y=df1['loan'] 
```

```python
#將資料切割成訓練集(train)及測試集(test)，決定切割比例test_size=0.25，隨機抽取第一筆資料的起始值random_state=5
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

```python
#利用邏輯斯迴歸分析建立預測模式
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```python
#計算正確率、精準度、召回率及綜合指標f1(train訓練集)
from sklearn.metrics import classification_report,confusion_matrix
LR_pred= LR.predict(X_train)
print(confusion_matrix(y_train,LR_pred))
print("======================================================\n")
print(classification_report(y_train,LR_pred))
```

```python
#計算正確率、精準度、召回率及綜合指標f1(test測試集)
from sklearn.metrics import classification_report,confusion_matrix
LR_pred= LR.predict(X_test)
print(confusion_matrix(y_test,LR_pred))
print("======================================================\n")
print(classification_report(y_test,LR_pred))
```

# 切割值
利用切割值p_value決定目標變數的類別

```python
#從LR預測模型中取得Y=0及Y=1的預測機率值(pre_test_prob)
pre_test_prob=pd.DataFrame(LR.predict_proba(X_test))
pre_test_prob
```

```python
#以切割值調整精準度,p為切割值，取得預測結果
p=0.2
pre_test=pre_test_prob[0].apply(lambda x: 0 if x>p else 1)
pre_test
```

```python
#計算混淆矩陣
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,pre_test))
```

```python
#評估預測結果的精準度、召回率、F1
from sklearn import metrics
print(metrics.classification_report(y_test,pre_test))
```
