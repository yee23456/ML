Regression Analysis(邏輯斯迴歸分析)方法」，以H/R,W/E,1Q,2Q,3Q,3P, TOT,OR,DR, A,ST,TO等特徵作為預測變數，建立球隊勝負之預測模式，並說明分析所得之主要結果。

(資料檔案：nba2.xlsx)
變數名稱：
Dateset: Regular Season
H/R: 主客場(1:Home; 2:Road)；W/E: 東西區(1:East; 2:West)
1Q,2Q,3Q,4Q: 第一、二、三、四節得分
PTS: points(得分)；3P: 三分球命中數；TOT: 籃板數
OR: 進攻籃板數；DR: 防守籃板數；A: 助攻數
ST: 抄截數；TO: 失誤數；W/L: 勝負(1:Win; 0:Loss)

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LogisticRegression
import statsmodels.api as smf
data = pd.read_excel('nba2.xlsx')
```

```python
X=data[['H/R','W/E','1Q','2Q','3Q','3P','TOT','OR','DR', 'A','ST','TO']]#輸入自變數/特徵
y=data['W/L']
```

#使用sklearn的LogisticRegression 模型

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```
