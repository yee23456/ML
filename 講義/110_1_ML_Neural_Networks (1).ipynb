```
import pandas as pd
data=pd.read_excel('loan3(na).xlsx')
```

```
data.head()
```

```
data.info()
```

```
df1=data.dropna()
X=df1[['Industry_type','Company_type','Beginning_time','Bank_relationship','Capital_increase','Age','Education','Credit']]
y=df1['loan'] 
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5,stratify=df1['loan'])
```

# **分類器**

```
#solver='lbfgs'：利用牛頓法求解，默認為梯度下降法；hidden_layer_sizes=(5,3)表示隱藏層有兩層第一層為八個神經元，第二層為三個神經元
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(8,3), random_state=1)
clf.fit(X_train,y_train)   
```

```
clf_pred= clf.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,clf_pred))
```

```
#計算混淆矩陣(confusion matrix)
print(confusion_matrix(y_test,clf_pred))
```

# **數量預測：迴歸器**

```
import pandas as pd
data2=pd.read_excel('boston_house.xlsx')
```

```
data2.head()
```

```
X=data2[['CRIM','INDUS','NOX','DIS','AGE','RAD','TAX','ZN','RM','PTRATIO','B','LSTAT']]
y=data2['PRICE'] 
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

```
#hidden_layer_sizes=(5,3)表示隱藏層有一層第一層為十二個神經元，設定激活函數為relu，默認即為relu,max_iter:迭代次數，默認：200
from sklearn.neural_network import MLPRegressor
reg = MLPRegressor(random_state=1, hidden_layer_sizes=(12), activation='relu',max_iter=10000)
reg.fit(X_train, y_train)
```

```
from sklearn.metrics import mean_squared_error
from math import sqrt
# 使用X_test預測y，得到y_pred1 
y_pred1 = reg.predict(X_test)
rmse = sqrt(mean_squared_error(y_test, y_pred1)) 
print("rms: %.3f" % rmse) 
```
