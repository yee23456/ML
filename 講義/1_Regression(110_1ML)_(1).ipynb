# 一、線性迴歸模式：解釋分析模式
用於分析哪些自變數(因素)會影響到某個依變數之統計分析模式，其中依變數需為“屬量”的變數，自變數亦為屬量變數

    可參考https://www.statsmodels.org/stable/api.html

```
X = df[["education_years","working_years"]] #education_years, working_years為自變數,可自行調整
y = df['salary'] 
```

```
import statsmodels.api as smf
RA = smf.OLS(y,X).fit()
print(RA.summary())
```

```
#進行迴歸分析，跑出迴歸分析解釋分析表(從statsmodels.formula.api 匯入ols線性迴歸分析)
#PTS為依變數；A, OEFF,TO為自變數，進行迴歸分析
import statsmodels.formula.api as smf
RA = smf.ols("salary ~ education_years+ working_years", df).fit()
print(RA.summary())
```

# 二、線性迴歸模式：預測分析模式_機器學習
利用線性迴歸模型，建立依變數之預測模型

```
#2.用線性回歸做訓練線性模型:
#import基本函式庫
import pandas as pd
import numpy as np

#準備X & y array
X = df[["education_years","working_years"]] #education_years, working_years為自變數,可自行調整
y = df['salary']            #salary為依變數,可自行調整

#將資料分成訓練組及測試組
from sklearn.model_selection import train_test_split

#test_size代表測試組比例。random_state代表設定隨機種子，讓測試結果可被重複
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=333)

#載入線性迴歸，並訓練模型
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)

#取得截距。如果公式是y=a+bx，a即是截距
print(lm.intercept_)
print(lm.coef_)
```

```
#取得迴歸係數，並用Data Frame顯示:pd.DataFrame(資料,列的名稱,行的名稱)
#lm.coef_
#X_train.columns
cdf = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coef'])
cdf
```

計算模式績效:RMSE, R_Square

```
#計算RMSE (可改用from sklearn.metrics import mean_squared_error, r2_score)
MSE1= np.mean((y_train - lm.predict(X_train) )** 2)
RMSE1=MSE1**0.5
MSE2= np.mean((y_test - lm.predict(X_test) )** 2)
RMSE2=MSE2**0.5
print('train_RMSE=%.3f,test_RMSE=%.3f'%(RMSE1,RMSE2))
```

```
#計算R_Square,
R2_train = lm.score(X_train,y_train)
R2_test = lm.score(X_test,y_test)

print('train_R2=%.3f, test_R2=%.3f'%(R2_train,R2_test))
```

```
df['salary'].mean()
```

```
from sklearn.metrics import mean_squared_error
from math import sqrt
# 使用X_test預測y，得到y_pred1 
y_pred1 = lm.predict(X_test)
rms = sqrt(mean_squared_error(y_test, y_pred1)) 
print("rms: %.3f" % rms) 
```

```
#輸入X1,X2, 計算模式所得的y值
lm.predict([[12,15],[8,10]])
```
